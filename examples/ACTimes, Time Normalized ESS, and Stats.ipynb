```python
from bounce import state
from bounce import driver
import numpy as np
```

```python
label = 'Emcee package test'
N_walkers = 20
N_iterations = 450*N_walkers
emcee_init_scale = {"m":0.00001, "a":.01, "h":0.001, "k":0.001, "l":0.01*2.*np.pi}
```

```python
initial_state = state.State(planets=[{"m":0.001, "a":0.1, "h":0.1, "k":0.15, "l":1.0}])
initial_state.hillRadiusFactor = 1.

np.random.seed(2018)
obs = driver.create_obs(initial_state, 100, 5.0e-4, 5.0e-5, 1.0)
```

```python
emcee_bundle1, emcee_bundle_hash1 = driver.run_emcee(label+"Niter", 
                                                   N_iterations, 
                                                   initial_state, 
                                                   obs,
                                                   N_walkers,
                                                   emcee_init_scale,
                                                   printing_every=100)

emcee_bundle2, emcee_bundle_hash2 = driver.run_emcee(label+"Niter*5", 
                                                   N_iterations*5, 
                                                   initial_state, 
                                                   obs,
                                                   N_walkers,
                                                   emcee_init_scale,
                                                   printing_every=100)
emcee_bundle3, emcee_bundle_hash3 = driver.run_emcee(label+"Niter*5*2", 
                                                   N_iterations*5*2, 
                                                   initial_state, 
                                                   obs,
                                                   N_walkers,
                                                   emcee_init_scale,
                                                   printing_every=100)
```

```python
driver.return_trimmed_results(emcee_bundle1, 40, [12,4], 0.7, plotting=False)
driver.return_trimmed_results(emcee_bundle2, 40, [12,4], 0.7, plotting=False)
driver.return_trimmed_results(emcee_bundle3, 40, [12,4], 0.7, plotting=False)
```

```python
driver.plot_ACTimes(emcee_bundle3, [12,4], 0.2, plotting=True)
```

```python
print driver.gelmanrubin_statistic(20, emcee_bundle1)
print driver.gelmanrubin_statistic(20, emcee_bundle2)
print driver.gelmanrubin_statistic(20, emcee_bundle3)
```
